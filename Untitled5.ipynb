2) How does broadcasting work in NumPy


Ans:-Broadcasting in NumPy is a powerful feature that allows arithmetic operations on arrays of different shapes without making unnecessary copies of data. It saves memory and simplifies your code.

üéØ What is Broadcasting?
Broadcasting is a set of rules NumPy uses to match the shapes of arrays during element-wise operations when the shapes differ but are compatible.

üìè Broadcasting Rules (Simplified)
When operating on two arrays, NumPy compares their shapes element-wise from right to left:

If the dimensions are equal, they‚Äôre compatible.

If one of the dimensions is 1, it can be stretched to match the other.

If the dimensions are different and not 1, it's an error.

‚úÖ Examples
1. Array + Scalar (Always Broadcasts)


```python
import numpy as np
a = np.array([1, 2, 3])
b = 10
print(a + b)  # [11 12 13]
```

Scalar 10 is "broadcast" to match the shape of a.

2. Array (3x1) + Array (1x3)


```python
a = np.array([[1], [2], [3]])  # Shape: (3, 1)
b = np.array([10, 20, 30])     # Shape: (1, 3)
result = a + b  # Shape: (3, 3)
print(result)
```

a is stretched across columns, b across rows.

3. Incompatible Shapes

```python
a = np.array([1, 2, 3])      # Shape: (3,)
b = np.array([[1, 2], [3, 4]])  # Shape: (2, 2)

# This will raise a ValueError
a + b  # ‚ùå Cannot broadcast (3,) and (2,2)
```

üß† Think of groupby() Like This:

```python
import pandas as pd

data = {
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [50000, 60000, 70000, 52000, 65000]
}

df = pd.DataFrame(data)
```

8)  What does the term ‚Äúvectorized operation‚Äù mean in NumPy?

Ans:- In NumPy, a vectorized operation refers to performing element-wise operations on entire arrays without using explicit loops (like for or while). Instead, operations are applied directly to arrays, leveraging under-the-hood optimizations in C for speed and efficiency.

üß† In Simple Terms:
Vectorization means doing operations all at once, instead of one element at a time.

‚úÖ Why Vectorized Operations Are Better
Feature	Benefit
üöÄ Faster	Uses optimized C loops internally (much faster than Python loops)
üßº Cleaner Code	More readable, less boilerplate
üß† More Pythonic	Feels like doing math with whole datasets
üßÆ Works like linear algebra	Natural for scientific and ML tasks

üìä Example: Traditional Loop vsVectorized
‚ùå Using a Python for loop:


```python
a = [1, 2, 3]
b = [4, 5, 6]
result = [a[i] + b[i] for i in range(len(a))]
```

‚úÖ Using NumPy vectorized operation:

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a + b  # Vectorized addition
```

üß™ Other Vectorized Operations in NumPy



```python
# Multiply
a * 2          # [2 4 6]

# Square root
np.sqrt(a)     # [1. 1.41 1.73]

# Element-wise comparison
a > 2          # [False False  True]

# Matrix multiplication (not element-wise)
np.dot(a, b)   # 1*4 + 2*5 + 3*6 = 32
```

üîç Why Hierarchical Indexing Is Important

3. ‚úÖ Advanced Slicing


```python
df.loc[('Sales', 'Q1')]  # Access specific group
df.loc['Sales']          # Access all quarters for Sales
```

4. ‚úÖ Reshaping with .unstack() or .stack()


```python
df.unstack()  # Moves inner level to columns
df.stack()    # Reverses the unstack
```

15) How does NumPy handle multidimensional arrays?

Ans:- NumPy handles multidimensional arrays very efficiently through its core data structure called an ndarray (n-dimensional array). Here‚Äôs how it works:

üîç Multidimensional Arrays in NumPy
A NumPy array (ndarray) can have any number of dimensions ‚Äî from 1D (vectors) to 2D (matrices), 3D, and beyond.

The shape of the array is represented as a tuple of integers, where each integer corresponds to the size along that dimension.

For example:

Shape (3,) = 1D array with 3 elements.

Shape (2, 4) = 2D array with 2 rows and 4 columns.

Shape (3, 2, 5) = 3D array with 3 blocks, each 2x5.

üß† Key Features
Feature	Explanation
Storage	Data stored in contiguous memory blocks for speed
Shape attribute	Accessed via .shape to see array dimensions
Indexing/Slicing	Supports multi-axis indexing like arr[0, 1] or arr[:, 2]
Broadcasting	Enables operations between arrays of different shapes
Vectorized operations	Efficient math applied across all elements in any dimension

üìã Example: Creating and Accessing Multidimensional Arrays

```python
import numpy as np

# Create a 2D array (matrix)
arr_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])

print(arr_2d.shape)  # Output: (2, 3)

# Access element in 2nd row, 3rd column
print(arr_2d[1, 2])  # Output: 6

# Create a 3D array
arr_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(arr_3d.shape)  # Output: (2, 2, 2)

# Access element
print(arr_3d[1, 0, 1])  # Output: 6
```

17)  Explain the difference between apply() and map() in Pandas?

Ans:- Great question! Both apply() and map() in Pandas are used to apply functions to data, but they differ in where and how they‚Äôre used.

üîç Differences Between apply() and map()
Aspect	map()	apply()
Used on	Primarily on Series	On both Series and DataFrame
Functionality	Maps values in a Series to new values using: function, dict, or Series	Applies a function along an axis (rows or columns) of DataFrame or on Series
Scope	Element-wise transformation	Element-wise or row/column-wise transformation
Input	Function, dictionary, or Series	Function that operates on Series, DataFrame row, or DataFrame column
Returns	Series with mapped values	Series or DataFrame depending on context
Use case examples	Replace values, map categories, simple transformations	More complex transformations involving rows/columns or entire Series

üß™ Examples
Using map() on a Series


```python
import pandas as pd

s = pd.Series(['cat', 'dog', 'rabbit'])
mapping = {'cat': 'kitten', 'dog': 'puppy'}

print(s.map(mapping))
# Output:
# 0    kitten
# 1     puppy
# 2       NaN
# dtype: object
```

Using apply() on a Series


```python
s = pd.Series([1, 2, 3, 4])

print(s.apply(lambda x: x ** 2))
# Output:
# 0     1
# 1     4
# 2     9
# 3    16
# dtype: int64
```

Using apply() on a DataFrame (row-wise)


```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

print(df.apply(lambda row: row['A'] + row['B'], axis=1))
# Output:
# 0    5
# 1    7
# 2    9
# dtype: int64
```

21) Why is NumPy‚Äôs array slicing faster than Python‚Äôs list slicing?


Ans:- NumPy's array slicing is significantly faster than Python's list slicing due to how NumPy stores and accesses data in memory. Here‚Äôs a breakdown of why:

‚ö°Ô∏è 1. NumPy Arrays Use Contiguous Memory Blocks
NumPy arrays are stored in a single, contiguous block of memory, unlike Python lists which are arrays of pointers to objects.

This memory layout allows NumPy to use efficient, low-level C routines to access and manipulate data, including slicing.

‚öôÔ∏è 2. Slicing Is a View, Not a Copy
NumPy slicing returns a view of the original data ‚Äî meaning no data is copied; only a new "window" on the same memory block is created.

In contrast, Python list slicing creates a new list with copied data, which takes more time and memory


```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
b = a[1:4]  # This is a view, not a new array
b[0] = 99
print(a)  # Output: [1 99 3 4 5]
```

üß† 3. Vectorized C Implementation
NumPy slicing is backed by compiled C code, which is optimized and much faster than Python‚Äôs interpreted loop-based operations.

The operations avoid Python-level overhead like reference counting, type checking, and dynamic dispatch.

üìã 4. Fixed Data Type (dtype)
NumPy arrays have a uniform data type, which allows for optimized memory access and arithmetic operations.

Python lists can hold mixed data types, so slicing has to account for more complexity and overhead.

‚è±Ô∏è Performance Comparison

```python
import numpy as np
import time

# NumPy slicing
arr = np.arange(1_000_000)
start = time.time()
arr_slice = arr[1000:200000]
end = time.time()
print("NumPy slice time:", end - start)

# Python list slicing
lst = list(range(1_000_000))
start = time.time()
lst_slice = lst[1000:200000]
end = time.time()
print("List slice time:", end - start)
```

‚û°Ô∏è You‚Äôll typically see NumPy slicing is faster ‚Äî and more memory-efficient ‚Äî especially with large datasets.
